```python
import pandas as pd
percorso_istat = "../../01_Data/Raw_Data/istat_incidenti_raw.csv"
percorso_situas = "../../01_Data/Raw_Data/Comuni - Dimensione Data Indagine 31-12-2020 Stampa 21062026185720.csv"
```

```python
#
df_istat = pd.read_csv(percorso_istat, dtype={'REF_AREA': str})
df_situas = pd.read_csv(percorso_situas, sep=';', dtype={'Codice Comune (alfanumerico)': str})
```

```python
colonne_utili_istat = ['REF_AREA', 'DATA_TYPE', 'RESULT', 'TIME_PERIOD', 'OBS_VALUE']

df_istat_clean = df_istat[colonne_utili_istat]

df_istat_clean = df_istat_clean.rename(columns={
    'REF_AREA': 'Codice_Comune',
    'DATA_TYPE': 'Tipo_Dato',
    'TIME_PERIOD': 'Anno',
    'OBS_VALUE': 'N. incidenti'
})

display(df_istat_clean.head(3))
```

```python
df_situas['Superficie (Kmq)'] = df_situas['Superficie (Kmq)'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)

colonne_utili_situas = [
    'Codice Comune (alfanumerico)', 
    'Comune', 
    'Popolazione residente', 
    'Superficie (Kmq)'
]
df_situas_clean = df_situas[colonne_utili_situas]

df_situas_clean = df_situas_clean.rename(columns={
    'Codice Comune (alfanumerico)': 'Codice_Comune',
    'Comune': 'Nome_Comune',
    'Popolazione residente': 'Popolazione',
    'Superficie (Kmq)': 'Superficie_Kmq'
})

display(df_situas_clean.head(3))
```

```python
df_merged = pd.merge(df_istat_clean, df_situas_clean, on='Codice_Comune', how='inner')
df_merged.shape[1]
```

```python
df_pivot = df_merged.pivot_table(
    index=['Codice_Comune', 'Nome_Comune', 'Popolazione', 'Superficie_Kmq', 'Anno'],
    columns='RESULT',
    values='N. incidenti',
    aggfunc='sum'
).reset_index()
```

```python
df_pivot = df_pivot.rename(columns={
    '9': 'Totale_Incidenti',
    'F': 'Feriti',
    'M': 'Morti'
})
```

```python
display(df_pivot.head())
```

```python
# Qua ho voluto eseguire un filtro temporale per selezionare solo gli anni che mi interessavano.
# Le giustificazioni sono: ----------------
anni_recenti = [2019, 2020, 2021, 2022, 2023]
df_finale = pd.DataFrame()
for anno in anni_recenti:
    df_temp = df_pivot[df_pivot['Anno'] == anno]
    df_finale = pd.concat([df_finale, df_temp])

df_finale.index = pd.RangeIndex(len(df_finale.index))
df_finale = df_finale.dropna().copy()

display(df_finale.head())
```

```python
# Ho voluto anche se non ancora del tutto puliiti, esportare questi dati con il filtro temporale per
# fare alcune verifiche di controllo. (scommentare per eseguire)
# percorso_export = "../../01_Data/Clean_Data/istat_filtrato_2019_2023.csv"
# df_finale.to_csv(percorso_export, index=False)
```

```python
df_finale['Tasso_Incidenti_10k'] = (df_finale['Totale_Incidenti'] / df_finale['Popolazione']) * 10000
df_finale['Densita_Incidenti_Kmq'] = df_finale['Totale_Incidenti'] / df_finale['Superficie_Kmq']

df_finale['Tasso_Incidenti_10k'] = df_finale['Tasso_Incidenti_10k'].round(2)
df_finale['Densita_Incidenti_Kmq'] = df_finale['Densita_Incidenti_Kmq'].round(2)

display(df_finale.head())
```

```python
# Qua ho deciso di aggiungere a posteriori una nuova colonna dove indica l'indice di fatalità,
# dopo durante l'analisi EDA mi è venuto in mente di aggiungere un nuovo indice di valutazione.
# Quindi ho dovuto calcolare l'indice di gravità in percentuale (% di incidenti con morti sul totale).

df_finale['Indice_Gravita_Perc'] = (df_finale['Morti'] / df_finale['Totale_Incidenti']) * 100

# Per i comuni con 0 incidenti il valore era NaN ma per il fare il calcolo giusto, i valori nan
# non ci devono essere, quindi li ho trasformati in 0.
df_finale['Indice_Gravita_Perc'] = df_finale['Indice_Gravita_Perc'].fillna(0)

# Qua ho voluto arrotondare a 2 decimali.
df_finale['Indice_Gravita_Perc'] = df_finale['Indice_Gravita_Perc'].round(2)

# Volevo verificare facendo una display.
display(df_finale.head(50))
```

```python
percorso_export_finale = "../../01_Data/Clean_Data/istat_incidenti_pulito_2019_2023.csv"
df_finale.to_csv(percorso_export_finale, index=False)
```
