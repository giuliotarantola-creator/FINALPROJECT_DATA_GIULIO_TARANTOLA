```python
import pandas as pd
percorso_istat = "../../01_Data/Raw_Data/istat_incidenti_raw.csv"
percorso_situas = "../../01_Data/Raw_Data/Comuni - Dimensione Data Indagine 31-12-2020 Stampa 21062026185720.csv"
```

```python
#
df_istat = pd.read_csv(percorso_istat, dtype={'REF_AREA': str})
df_situas = pd.read_csv(percorso_situas, sep=';', dtype={'Codice Comune (alfanumerico)': str})
```

```python
colonne_utili_istat = ['DATAFLOW', 'FREQ', 'REF_AREA', 'DATA_TYPE', 'RESULT', 'TIME_PERIOD', 'OBS_VALUE']

df_istat_clean = df_istat[colonne_utili_istat]

df_istat_clean = df_istat_clean.rename(columns={
    'REF_AREA': 'Codice_Comune',
    'DATA_TYPE': 'Tipo_Dato',
    'TIME_PERIOD': 'Anno',
    'OBS_VALUE': 'Valore'
})

display(df_istat_clean.head(3))
```

```python
df_situas['Superficie (Kmq)'] = df_situas['Superficie (Kmq)'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)

colonne_utili_situas = [
    'Codice Comune (alfanumerico)', 
    'Comune', 
    'Popolazione residente', 
    'Superficie (Kmq)'
]
df_situas_clean = df_situas[colonne_utili_situas]

df_situas_clean = df_situas_clean.rename(columns={
    'Codice Comune (alfanumerico)': 'Codice_Comune',
    'Comune': 'Nome_Comune',
    'Popolazione residente': 'Popolazione',
    'Superficie (Kmq)': 'Superficie_Kmq'
})

display(df_situas_clean.head(3))
```
